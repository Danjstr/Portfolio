**Models and Statistical Analysis**

##### Let's build a linear model to examine if there is a linear relationship between total cases per million and population density. 

```python
import numpy as np

# use a linear model
from sklearn.linear_model import LinearRegression

# provide data
x1 = np.array(df3['aged_70_older']).reshape((-1, 1))
y1 = np.array(df3['total_cases_per_million'])

# create a model and fit it
model = LinearRegression()
model.fit(x1, y1)
model = LinearRegression().fit(x1, y1)

# calculate r-sqr value
r_sq1 = model.score(x1, y1)
print('coefficient of determination:', r_sq1)
```

```python
sns.regplot(x1,y1)
plt.ylabel('aged_70_older')
plt.xlabel("total_cases_per_million")
plt.title("Relationship Between People Age 70 and older and Cases Per Million")
```

##### The r-square value is 0.00088, which is relatively small, means only 0.08% of provided data can fit the model, so we say that total cases per million is not necessarily associated with population density. 
##### //
##### Let's generate a hypothesis test using new smoothed cases and gdp to see if there is a relationship between economical condition and smoothing proportion. 
##### Null hypothesis: There is a linear relationship between new smoothed cases and gdp. 
##### Alternative hypothesis: There is no linear relationship between new smoothed cases and gdp.

```python
# provide data
x2 = np.array(df3['gdp_per_capita']).reshape((-1, 1))
y2 = np.array(df3['new_cases_smoothed_per_million'])

# build a linear model
from sklearn import linear_model
from regressors import stats
ols = linear_model.LinearRegression()
ols.fit(x2, y2)

print('coefficient of determination:', ols)
# get summary table
stats.summary(ols, x2, y2)
```

```python
sns.regplot(x2,y2)
plt.xlabel("DGP Per Capita")
plt.ylabel("New_cases_per_million")
plt.title("Relationship between GDP and Cases Per Million")
```

##### We have a 0 p-value so we can reject the null hypothesis, but since the r-square value is relatively low, we say that there is a slight linear relationship between new smoothed cases per million and gdp. 

paired sample t-test

*   H0: There is no difference on the amount of new cases in North America between 3/1/2021 and 4/1/2021. 
*   H1: There is a difference on the amount of new cases in North America between 3/1/2021 and 4/1/2021. 



```python
dfna = df1.loc[df1['continent'] == 'North America']
```

```python
s1 = dfna.loc[dfna['date'] == '2021-03-01', ['new_cases']].values
s2 = dfna.loc[dfna['date'] == '2021-04-01', ['new_cases']].values

import numpy as np
nattest = np.concatenate([s1, s2], axis=1)
nattest = pd.DataFrame(nattest, columns=['210301','210401'])
```

```python
import pandas as pd
from scipy import stats

nattest[['210301','210401']].describe()
nattest,pval = stats.ttest_rel(nattest['210301'], nattest['210401'])
print(pval)
if pval<0.05:
    print("reject null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

paired sample t-test

* H0: There is no difference on the amount of new deaths in North America 
between 3/1/2021 and 4/1/2021.
* H1: There is a difference on the amount of new deaths in North America between 3/1/2021 and 4/1/2021.

```python
s3 = dfna.loc[dfna['date'] == '2021-03-01', ['new_deaths']].values
s4 = dfna.loc[dfna['date'] == '2021-04-01', ['new_deaths']].values

import numpy as np
nattest1 = np.concatenate([s3, s4], axis=1)
nattest1 = pd.DataFrame(nattest1, columns=['210301','210401'])

import pandas as pd
from scipy import stats

nattest1[['210301','210401']].describe()
nattest1,pval1 = stats.ttest_rel(nattest1['210301'], nattest1['210401'])
print(pval1)
if pval1<0.05:
    print("reject null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

```python
X = df3['new_cases_smoothed_per_million'].values
result = adfuller(X)
print('ADF Statistic: %f' % result[0])
print('p-value: %f' % result[1])
print('Critical Values:')
for key, value in result[4].items():
    print('\t%s: %.3f' % (key, value))

if result[0] < result[4]["5%"]:
    print ("Reject Ho - Time Series is Stationary")
else:
    print ("Failed to Reject Ho - Time Series is Non-Stationary")
```

<center><h1>Auto Regressive Model</h1></center>

$$
Y_{t}=\alpha+\beta_{1} Y_{t-1}+\beta_{2} Y_{t-2}+\ldots+\beta_{p} Y_{t-p}+\epsilon_{1}
$$

```python
plot_pacf(X, lags=10);
plt.ylabel("correlationship")
plt.xlabel("Number of Lags")
```

```python
df3['new_cases_per_million_shifted']  = df3['new_cases_smoothed_per_million'].shift()
```

```python
df3.head()
```

```python
df5 = df3.drop(df3.columns.difference(['new_cases_smoothed_per_million','new_cases_per_million_shifted']), 1)
```

```python
df5['new_cases_per_million_shifted']=df5['new_cases_per_million_shifted'].fillna(0)
```

```python
df5.head()
```

```python
y = df5['new_cases_smoothed_per_million'].values
X = df5['new_cases_per_million_shifted'].values
```

```python
train_size = int(len(X) * 0.80)
```

```python
X_train, X_test = X[0:train_size], X[train_size:len(X)]
y_train, y_test = y[0:train_size], y[train_size:len(X)]
```

```python
X_train = X_train.reshape(-1,1)
X_test = X_test.reshape(-1,1)
```

```python
lr = LinearRegression()
lr.fit(X_train, y_train)
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
y_pred = lr.predict(X_test)
```

```python
rcParams["figure.figsize"] =10, 10
plt.plot(y_test[-100:], label="Actual Values")
plt.plot(y_pred[-100:], label="Predicted Values")
plt.xlabel("Predicted Records")
plt.ylabel("Predicted Cases Per Record")
plt.title(" Predicted Cases Per Million")
plt.legend()
plt.show()
```

```python
model = ARIMA(y_train, order=(1,0,0))
```
